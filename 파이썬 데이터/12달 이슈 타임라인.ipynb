```python
import os
import time
import datetime
from konlpy.tag import Kkma,Okt
from konlpy.utils import pprint
from collections import Counter
from wordcloud import WordCloud
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt 



current_path = os.getcwd()+"/data"

dirlist=os.listdir(current_path)
year=input('분석하고 싶은 연도를 입력하세요: ')

noun_list=[]
word=[]
frequency=[]
for i in range(1,13):
    if i <10:
        cnt= '0'+str(i)
    else :
        cnt=str(i)

    f=open(current_path+"/"+"경제"+year+cnt+".txt","r")
    data= f.read()
    
    
    #Okt 객체 생성
    okt = Okt()
    noun= okt. nouns(data)

    #필요없는 단어 제거
    pop_list=["연합뉴스","통해","위해","크게","여기","최근","내년","수준","현재","이번","대한","분야","재배포","전재","무단","금지","제공","배포","매일 경제","기자"]
    noun = [word for word in noun if len(word) >= 2 and word not in pop_list]
    count= Counter(noun)

#명사 빈도 카운트 가장 많이 나온 단어 1개
  
    for n,c in count.most_common(1):
        temp={'tag':n,'count':c}
        noun_list.append(temp)


    #명사와 빈도수 따로 추출

for tag in noun_list:
    noun=tag['tag']
    count=tag['count']
    word.append(noun)
    frequency.append(count)

 #현재 작업 디렉토리
current_path = os.getcwd()
# print(current_path)
#변경 디렉토리
os.chdir('C:/Users/big1-01/Desktop/project/renewkdi2/renewKDI/src/main/resources/static/images/bigdata')


#화면에 그래프로 출력

plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10,10))

plt.rc('font', size=20) 
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=15)

title_font = {
    'fontsize': 24,
    'fontweight': 'bold'
}

titl=['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']  
# print(titl)
for i in range(0,12): 
    plt.text(titl[i],frequency[i],word[i],fontsize=15, color='black',
             
            horizontalalignment ='center')
plt.bar(titl,frequency,color='skyblue')
plt.title(year+'년도 이슈 타임라인', fontdict=title_font)
plt.savefig('경제이슈타임라인.png', bbox_inches='tight')

plt.show()



    

    

    
```
