```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import os

#  
current_path = os.getcwd()
print(current_path)
# 
os.chdir('C:/Users/big1-08/git/renewkdi2/renewKDI/src/main/resources/static/images')



df= pd.read_csv('.csv',encoding='cp949')
teen=[]
for i in range(7):
    teen.append(df.loc[i].to_list())
label=["  ","","  ","  ","","&"]
explode = [0, 0, 0.10, 0,0,0]
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
for i in range(7):
    plt.rc('font',family='Malgun Gothic')
    plt.axis('equal')
    plt.pie(teen[i][1:],labels=label,autopct='%.1f%%',textprops={'fontsize': 10}, startangle=260, explode=explode,shadow=True,wedgeprops=wedgeprops)
    plt.savefig(''+str(i)+'.png')
    plt.show()
```
