```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import os

#현재 작업 디렉토리
current_path = os.getcwd()
print(current_path)
#변경 디렉토리
os.chdir('C:/Users/big1-08/git/renewkdi2/renewKDI/src/main/resources/static/images')



df= pd.read_csv('양극화불평등지수.csv',encoding='cp949')
teen=[]
for i in range(7):
    teen.append(df.loc[i].to_list())
label=["변화하는 가족 구조","대인관계갈등","사회 양극화와 갈등","실업과 직장 스트레스","정신건강","지역불균현&문화격차"]
explode = [0, 0, 0.10, 0,0,0]
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
for i in range(7):
    plt.rc('font',family='Malgun Gothic')
    plt.axis('equal')
    plt.pie(teen[i][1:],labels=label,autopct='%.1f%%',textprops={'fontsize': 10}, startangle=260, explode=explode,shadow=True,wedgeprops=wedgeprops)
    plt.savefig('양극화불평등지수'+str(i)+'.png')
    plt.show()
```
