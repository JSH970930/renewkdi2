## epu지수

```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import os

#현재 작업 디렉토리
current_path = os.getcwd()
print(current_path)
#변경 디렉토리
os.chdir('C:/Users/big1-08/git/renewkdi2/renewKDI/src/main/resources/static/images')

f = open('epu.csv', 'r')
data = csv.reader(f)
next(data)
epu =[]
date=[]

for row in data:
    epu.append(float(row[1]))
    date.append(row[0])

xdate=[xdate[0:4] for xdate in date]    

font_path = "C:/windows/Fonts/batang.ttc"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

plt.figure(figsize=(40,21))
plt.rc('font', size=37)


ax = plt.subplot()
ax.margins(x=0)

ax.patch.set_facecolor('lightgrey')
ax.patch.set_alpha(0.5)

ax.grid(True)

plt.plot(date, epu,'ko-',markersize=10)
plt.xticks(ticks=date, labels=xdate)
plt.locator_params(axis='x', nbins=len(xdate)/12)
plt.savefig('epu.png')
plt.show()

```

## 연령별 고용불안 지수

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import os

#현재 작업 디렉토리
current_path = os.getcwd()
print(current_path)
#변경 디렉토리
os.chdir('C:/Users/big1-08/git/renewkdi2/renewKDI/src/main/resources/static/images')



df= pd.read_csv('연령별고용불안지수.csv',encoding='cp949')
teen=[]
for i in range(7):
    teen.append(df.loc[i].to_list())
label=["매우느낌","느끼는편임","느끼지않음","전혀느끼지않음"]
explode = [0, 0.10, 0, 0.10]
color = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
for i in range(7):
    plt.rc('font',family='Malgun Gothic')
    plt.axis('equal')
    plt.pie(teen[i][1:],labels=label,autopct='%.1f%%',textprops={'fontsize': 10}, startangle=260, explode=explode,shadow=True,colors=color, wedgeprops=wedgeprops)
    plt.savefig('고용불안지수'+str(i)+'.png')
    plt.show()
    
```
