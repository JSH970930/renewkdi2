```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import os

#현재 작업 디렉토리
current_path = os.getcwd()
print(current_path)
#변경 디렉토리
os.chdir('C:/Users/big1-08/git/renewkdi2/renewKDI/src/main/resources/static/images')

f = open('행복지수.csv', 'r')
data = csv.reader(f)

x1=[]
y1=[]
for row in data:
    row[2]=float(row[2])
    x1.append(row[0]+'위'+' '+row[1])
    y1.append(float(row[2]))

# print(x1)
# print(y1)
 
x=x1[:5]
x.extend(['...'])
x.extend(x1[len(x1)-3:len(x1)])
print(x)
y=y1[:5]
y.extend([0])
y.extend(y1[len(y1)-3:len(y1)])
plt.figure(figsize=(15,10))
plt.rc('font', family='Malgun Gothic')
plt.rc('font', size=15)
print(y)
plt.barh(x, y)
plt.ylim([-0.5,6.5])
plt.gca().invert_yaxis()
plt.savefig('행복지수.png')

plt.show()
```
